```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/titanic.csv")
df.tail(10)
```

```python
def bar_char(feature):  # Se corrigió 'future' a 'feature'
    survived = df[df["Survived"] == 1][feature].value_counts()
    dead = df[df["Survived"] == 0][feature].value_counts()

    test = pd.DataFrame([survived, dead])
    test.index = ["Sobrevivientes", "Fallecidos"]

    test.plot(kind="bar", stacked=True, figsize=(10, 5))
    plt.title(f'Comparación de {feature} entre sobrevivientes y fallecidos')
    plt.xlabel(feature)
    plt.ylabel("Cantidad")
    plt.show()
```

```python
bar_char("Sex")
```

```python
bar_char("Pclass")
```

```python
bar_char("Embarked")
```

##Ingenieria de caracteristicas

Es el proceso de usar el dominio del conocimiento para crear nuevas características (feature vector) y que logre que los algoritmos de machine learning funcionen correctamente. Feature vector es un vecotr n-dimensional que representa un objeto. Muchos algoritmos de ML requieren una representación numérica de los objetos, ya que facilitan el procesamiento y el análisis estadístico.
