```python
data2 = pd.read_csv("/content/dataset_edad_precio_seguro.csv")
data2.head()
```

```python
model2 = LinearRegression()
X = data2["Edad"]
Y = data2["PrecioSeguro"]

model2.fit(X.values.reshape(-1, 1),Y)

y_pred2=model2.predict(X.values.reshape(-1, 1))

r2_seguro=r2_score(Y,y_pred2)

plt.figure(figsize=[10,6])
plt.scatter(X, Y, color = "green",label="Datos Reales")
plt.plot(X, y_pred2,color="purple",linewidth=2,label="Regresión lineal")
plt.xlabel("Edad")
plt.ylabel("Precio seguro")
plt.title("Precio del seguro vs edad")
plt.legend()
plt.grid(True)
plt.show

print(f"Ecuación de la recta: y = {model2.coef_[0]:.2f}x + {model2.intercept_:.2f}")
print(f"Coeficiente de determinación: {r2_seguro:.2f}")
```

```python
data3 = pd.read_csv("/content/house_prices_large.csv")
data3.head()
```

```python
X = data3["House_Size"]
Y = data3["Price"]
model3=LinearRegression()
model3.fit(X.values.reshape(-1, 1),Y)

y_pred3=model3.predict(X.values.reshape(-1, 1))

r2=r2_score(Y,y_pred3)

plt.figure(figsize=[10,6])
plt.scatter(X, Y, color = "green",label="Datos Reales")
plt.plot(X, y_pred3,color="purple",linewidth=2,label="Regresión lineal")
plt.xlabel("Tamaño de casa")
plt.ylabel("Precio de casa")
plt.title("Tamaño de casa vs precio de casa")
plt.legend()
plt.grid(True)
plt.show

print(f"Ecuación de la recta: y = {model3.coef_[0]:.2f}x + {model3.intercept_:.2f}")
print(f"Coeficiente de determinación: {r2:.2f}")
```
